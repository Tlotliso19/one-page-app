```python
from pathlib import Path

# Path to the file you want to read
file_path = Path("files/explorative.ipynb")

# List to store file details
file_objects = []

# Check if the file exists and is a file
if file_path.is_file():
    file_name = file_path.name  # Get the name of the file
    file_data = {'name': file_name, 'content': None}

    try:
        # Open and read the file content
        with open(file_path, 'r', encoding='utf-8') as f:
            file_data['content'] = f.read()  # Read the file content into a string
    except Exception as e:
        file_data['content'] = f"Error reading file: {e}"

    # Append the file data to the list
    file_objects.append(file_data)
else:
    print(f"File {file_path} does not exist or is not a valid file.")
```

```python
contents=[]
for object in file_objects:
    contents.append(object['content'])  
```

```python
from pathlib import Path

# Path to the file you want to read
file_path = Path("files/explorative.ipynb")

# List to store file details
file_objects = []

# Check if the file exists and is a file
if file_path.is_file():
    file_name = file_path.name  # Get the name of the file
    file_data = {'name': file_name, 'content': []}  # Store content as a list of rows

    try:
        # Open and read the file line by line (row by row)
        with open(file_path, 'r', encoding='utf-8') as f:
            # Read each line (row) and append it to the content list
            file_data['content'] = [line.strip() for line in f.readlines()]
           

    except Exception as e:
        file_data['content'] = f"Error reading file: {e}"

    # Append the file data to the list
    file_objects.append(file_data)
else:
    print(f"File {file_path} does not exist or is not a valid file.")

```

```python
contents=file_objects[0]['content']
```

```python
for content in contents:
    print('=====================================================')
    print(content)
```
